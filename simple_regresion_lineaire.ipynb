```python
#génération de données 
import numpy as np
X = np.array([0, 3, 6, 8])
Y = np.array([35, 45, 65, 80])
```

```python
import matplotlib.pyplot as plt
plt.plot(X, Y, '*') 
plt.xlabel("Années d'expérience | notre variable explicative 'x' ") 
plt.ylabel("Salaire | Variable cible 'y'") 
plt.title("Scratter plot") 
plt.savefig("./intuitive_scatter.png")
```

```python
def reg_plot(x, y, m):
    plt.scatter(x, y, c = 'blue', label = "Nos données")
    plt.plot(x, m.predict(x.reshape(-1, 1)), color = 'red', label = "Courbe de prédiction")
    plt.xlabel("Variable explicative 'x'")
    plt.ylabel("Variable cible 'y'")
    plt.legend()
    return None
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X.reshape(-1, 1), Y)
reg_plot(X, Y, linear_model)
plt.savefig( "./approche_intuitive.png")
```

```python
import numpy as np
# Génération des données
x = np.arange(75)
delta = np.random.uniform(-10, 10, size = (75,))
y = 0.4 * x + 3 + delta
```

```python
import pandas as pd  
df = pd.read_csv("./data/Salary_Data.csv", sep = ',') 
df.head ()
```

```python
df = df[["YearsExperience", "Salary"]]
```

```python
X = df.YearsExperience 
Y = df.Salary 
plt.plot(X , Y , '*')
plt.xlabel("Années d'expérience | notre variable explicative 'x' ")
plt.ylabel("Salaire | Variable cible 'y' ")
plt.title("Années d'expérience vs Salaire " )
```

```python
from sklearn.linear_model import LinearRegression 
linear_model = LinearRegression ()
linear_model.fit(np.array(X).reshape(-1, 1),np.array(Y))
reg_plot(np.array(X),np.array(Y),linear_model)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Prédictions du modèle sur les données de 'YearsExperience'
Y_pred = linear_model.predict(np.array(X).reshape(-1, 1))

# Calcul et affichage de l'erreur quadratique moyenne (MSE)
print("Mean squared error: %.2f" % mean_squared_error(Y, Y_pred))
# L'erreur quadratique moyenne (MSE) est la moyenne des carrés des écarts entre les valeurs observées (Y) et les valeurs prédites (Y_pred). Plus elle est faible, plus le modèle est précis.

# Calcul et affichage de la racine carrée de l'erreur quadratique moyenne (RMSE)
print("Root mean squared error: %.2f" % np.sqrt(mean_squared_error(Y, Y_pred)))
# La racine carrée de l'erreur quadratique moyenne (RMSE) est une mesure de l'écart type des résidus (les erreurs de prédiction). Elle est en général plus interprétable car elle est dans la même unité que la variable cible (ici, le salaire).

# Calcul et affichage du coefficient de détermination R^2
print("R square: %.2f" % r2_score(Y, Y_pred))
# Le R^2 est le coefficient de détermination qui mesure la proportion de la variance de la variable cible (Y) qui est expliquée par le modèle. Il varie entre 0 et 1 : plus il est proche de 1, plus le modèle est performant (1 signifie que le modèle explique 100 % de la variance de Y).
```
